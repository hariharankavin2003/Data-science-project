```python
import re #to remove white spaces,punchution lam
```

```python
def clean(text):                                #TO REMOVE WHITE SPPACES AND PUNCHUTION
    text = re.sub('[^A-Za-z]+', ' ',str(text))
    return text
```

```python
df2=df1.astype('object')
```

```python
df2['cleaned']=df2['Review'].apply(clean)
```

```python
df2
```

```python
df2['Lowercase']= df2['cleaned'].apply(lambda x: " ".join(word.lower() for word in x.split()))
```

```python
df2
```

```python
#################################################################################################
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
#stop_words.remove('i')  #to remove na
```

```python
df2['stopwords'] = df2['Lowercase'].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
df2
```

```python
#####################################################################################################
```

```python
#splitting the english words
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk import pos_tag
import nltk

# Download necessary NLTK data files
nltk.download('punkt')
nltk.download('wordnet')

# POS tagger dictionary
pos_dict = {'J': wordnet.ADJ, 'V': wordnet.VERB, 'N': wordnet.NOUN, 'R': wordnet.ADV}

def token_pos(text):
    return [
        (word, pos_dict.get(tag[0]))
        for word, tag in pos_tag(word_tokenize(text))
    ]

# Apply to your data
df2['POS tagged'] = df2['stopwords'].apply(token_pos)
```

```python
from nltk.stem import WordNetLemmatizer

# Initialize the lemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

def lemmatize(pos_data):
    lemma_rew = " "
    for word, pos in pos_data:
        if not pos:  # If no POS tag is available, use the word as is
            lemma = word
        else:
            # Lemmatize based on the POS tag
            lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
        lemma_rew = lemma_rew + " " + lemma  # Add the lemma to the result string
    return lemma_rew

# Assuming df2 is your dataframe and 'POS tagged' contains POS-tagged text
df2['Lemma'] = df2['POS tagged'].apply(lemmatize)

# Display the first few rows of the dataframe
df2.head()
```

```python
#####################################################################################################################3
```

```python
# Using TEXTBLOB to analyze the sentiment of the text by returning a polarity and subjectivity score.
#positive or negative nu soluum
import pandas as pd
from textblob import TextBlob

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0.1:  # Positive
        return 'Positive'
    elif polarity < -0.1:  # Negative
        return 'Negative'
    else:  # Neutral
        return 'Neutral'

# Apply the function to the 'Lemma' column (or 'Cleaned Reviews')
df2['sentiment'] = df2['Lemma'].apply(get_sentiment)

# Display the dataframe to check the sentiment column
df2.head(10)
```

```python
df2['sentiment'].unique()
```

```python
le=LabelEncoder()
```

```python
df2['sentiment']=le.fit_transform(df2['sentiment'])
```

```python
x=df2['Lemma']
```

```python
y=df2['sentiment']
```

```python
# Vectorize the text
vectorizer = CountVectorizer()
x_vectorized = vectorizer.fit_transform(x)

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x_vectorized, y, test_size=0.25, random_state=42)
```

```python
##########################################################################################################################3
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
y_pred=lg.predict(x_test)
```

```python
print(accuracy_score(y_pred,y_test)*100)
```
